```python
# getting country records , zomato is more available in india, max numbers of trns is from India

country_names= final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
#pie chart banao [:3]<- iska matlab top 3 laane ko tha to use kiye '%1.1f%%'<- isko lagaye taki pie pe value show ho
plt.pie(country_val[:2], labels = country_names[:2], autopct = '%1.1f%%')
```
